```python
import google.generativeai as genai
from google.colab import userdata

# Configuraci√≥n de la API
try:
  API_KEY = userdata.get('GOOGLE_API_KEY')
  genai.configure(api_key=API_KEY)
  MODEL_NAME = "gemini-3-flash-preview"
  print("API configurada.")
except Exception as e:
  print(f"Error de configuraci√≥n:{e}")
```

```python
model = genai.GenerativeModel(MODEL_NAME)

def generar_estrategia_pasteleria(tipo_tarea, contexto):
    """
    POC de Fast Prompting para IA-BOOST.
    Implementa l√≥gica de ventas y generaci√≥n de contenido.
    """

    # T√©cnica: Persona Prompting + Few-Shot
    instrucciones_base = """
    Act√∫a como un experto vendedor de pasteler√≠a artesanal premium.
    Tu tono es emp√°tico, dulce y muy profesional.
    No uses conservantes ni procesos industriales en tus argumentos.
    """

    prompts = {
        "objecion": f"Responde de forma corta a un cliente que dice que el precio es elevado. Producto: {contexto}",
        "copy_aida": f"Genera un copy de venta usando la t√©cnica AIDA (Atenci√≥n, Inter√©s, Deseo, Acci√≥n) para: {contexto}"
    }

    prompt_final = f"{instrucciones_base}\n\nTarea: {prompts[tipo_tarea]}"

    response = model.generate_content(prompt_final)
    return response.text
```

Ejempolos de uso


```python
# Ejemplo de uso 1: El cliente dice que la Torta Matilda es cara.
producto_premium = "Torta Matilda (capas intensas de chocolate y dulce de leche)"
resultado_objecion = generar_estrategia_pasteleria("objecion", producto_premium)

print("--- ESCENARIO: El cliente dice que es caro ---")
print(f"Producto: {producto_premium}")
print(f"Respuesta de la IA:\n{resultado_objecion}")
```

```python
# Ejemplo de uso 2: Crear un posteo para atraer ventas de media tarde.
producto_clasico = "Docena de Alfajores de Maicena cl√°sicos con mucho coco"
resultado_copy = generar_estrategia_pasteleria("copy_aida", producto_clasico)

print("--- ESCENARIO: Posteo para Redes Sociales ---")
print(f"Producto: {producto_clasico}")
print(f"Copy generado:\n{resultado_copy}")
```

```python
# Ejemplo de uso 3: Promocionar la variedad de budines.
producto_variado = "Budines artesanales (Vainilla, Chocolate o Marmolado, con opci√≥n de relleno de dulce de leche)"
resultado_budin = generar_estrategia_pasteleria("copy_aida", producto_variado)

print("--- ESCENARIO: Lanzamiento de variantes de Budines ---")
print(f"Producto: {producto_variado}")
print(f"Copy generado:\n{resultado_budin}")
```
